WAVE EQUATION: 2nd order PDE in space & time:

1D wave equation ("1 + 1") = one spacial dimension $x$ and temporal dimension $t$. 

$$
\frac{1}{v^2} \frac{\partial ^2 f}{\partial t^2} = \frac{\partial ^2 f}{\partial x^2}$$

$f = f(x, t)$, where $v $ = speed = $dx/dt$. 

But in 2D + 1 time: $f = f(x, y, t)$

$$ \frac{\partial ^2 f}{\partial x^2} + \frac{\partial ^2 f}{\partial y^2} = \frac{1}{v^2} \frac{\partial ^2 f}{\partial t^2} $$

DEF: A "wave" is anything that satisfies the wave equation. 

and in 3D: $f = f(x, y, z, t)$

$$ \frac{\partial ^2 f}{\partial x^2} + \frac{\partial ^2 f}{\partial y^2} + \frac{\partial ^2 f}{\partial z^2} = \frac{1}{v^2} \frac{\partial ^2 f}{\partial t^2} $$

A word or two on Helmholz Decomposition (Helmholz theorem as Griffiths calls it):

Q: How do you express a vector field in terms of only div and curl?
A: It's not the most straightforward, but

$\vec v = -\vec \nabla \phi + \vec \nabla \times \vec \xi$.

So...why?

Take div:

$$
\vec \nabla \cdot \vec v = \vec \nabla \cdot (- \vec \nabla \phi ) + \bcancel{\vec \nabla \cdot (\vec \nabla \times \vec \xi)}
$$

So divergence is entirely determined by $\phi$ (a scalar). 

Similarly when we look at curl:

$$\vec \nabla \times \vec v = \bcancel{\vec \nabla \times (-\vec \nabla \phi)} + \vec \nabla \times (\vec \nabla \times \vec \xi)$$

...which is determined only by $\xi$.

Now suppose $\vec \nabla \cdot \vec v = 0$. Then the field has only curl:

$$\vec v = \vec \nabla \times \vec \xi $$

ah! Now $\vec v$ is the curl of an "underlying" vector potential!

Famously, $\vec \nabla \cdot \vec B = 0$. So $\vec B$ can be written as the curl of some other field:

$$\boxed{\vec B = \vec \nabla \times \vec A}$$

$\vec B$ is derived from an underlying vector potential $\vec A$. We will revisit this later. 

What if, however, a field is curl-free? (irrotational)

$$\vec \nabla \times \vec v = 0 \implies \vec v = -\vec \nabla \phi$$
so $\vec v$ is the gradient of the scalar potential. What is that potential...?

Recall, in electrostatics, charges not moving in time, means $\vec B \neq 0$, means

$$\vec \nabla \times \vec E = 0 \quad \text{for \underline{statics.}}$$

So, $$\vec E = -\vec \nabla V$$

where $V$ is the electrostatic potential (measured in volts).

But, by examination, the $d\vec  a_\text{BACK}, \, d\vec a_\text{TOP}, \, d\vec a_\text{LEFT}, \text{and} \, d\vec a_\text{RIGHT}$ terms cancel, because $\vec E \parallel d\vec a$ for all four surfaces. 
$$
\oint \limits _{\text{G.S.}}\vec E \cdot d\vec a = \int\limits_\text{TOP} E_{\perp \text{ab}} \cdot da_{\text {ab}} \, - \int\limits_{\text{BOTTOM}}E_{\perp \text{below}} \cdot da_{\text {below}}
$$

So the equation is now:

$$
\begin{align*}
\oint \vec E \cdot d\vec a &= E_{\perp \text{ab}}\int da_\text{ABV}- E_{\perp \text{below}}\int da_\text{below} \\
&= E_{\perp \text{ab}}(A)- E_{\perp \text{below}}(A) \leftarrow \text{call that RHS}
\end{align*}
$$

So when we compare:

$$
A(E_{\perp \text{above}} - E_{\perp \text{below}}) = \frac{\sigma A}{\varepsilon_0} \implies \boxed{E_{\text{above}} - E_{\text{below}} = \frac{\sigma}{\varepsilon_0} \quad \text{EQ 2.31}}
$$

This means there is a discontinuity in the normal component of an electric field at any boundary. Referenced in the book: EQ. 2.35 - 2.37 expresses this in terms of $V$, $\hat n$, and $\frac{\partial V}{\partial n}$.

Separation of variables: start with separable ansatz. In 2D cartesian coords, the _separable ansatz_ is:

$$
V = X(x)Y(y). \quad \text{Now \boxed{\text{plug into}\, \nabla^2 \,\text{ansatz}}}
$$
$$
\nabla^2 V = 0 \implies \frac{\partial ^2 V}{\partial x^2} + \frac{\partial ^2 V}{\partial y^2} \implies \frac{\partial^2}{\partial x^2} \left(X(x)Y(y)\right)+\frac{\partial^2}{\partial y^2} \left(X(x)Y(y)\right) = 0
$$

$$
= Y(y) \frac{\partial ^2 X(x)}{\partial x^2} + X(x) \frac{\partial ^2 Y(y)}{\partial y^2}  = 0 \quad \boxed {\text{slide functions past partials}}
$$

$$
= \frac{\bcancel{Y(y)}}{X(x)\bcancel{Y(y)}} \frac{\partial ^2 X(x)}{\partial x^2} + \frac{\bcancel{X(x)}}{\bcancel{X(x)}Y(y)} \frac{\partial ^2 Y(y)}{\partial y^2} = 0 \quad \boxed {\text{divide by V}}
$$

$$
= \underbrace{\frac{{1}}{X(x)} \frac{d ^2 X}{d x^2}}_\text{all dep on x} + \underbrace{\frac{1}{Y(y)} \frac{d ^2 Y}{d y^2}}_{\text{all dep on y}} = 0 \quad \boxed {\text{replace partials w/ totals}}
$$
Insight #1: each term must seperately be equal to a constant
Insight #2: equal & opposite constants
Let's let

$$
\frac{{1}}{X(x)} \frac{d ^2 X}{d x^2} = k^2 \quad\text{and}\quad \frac{1}{Y(y)} \frac{d ^2 Y}{d y^2} = -k^2
$$
Recognizing famous D.E. solutions to which are known: "wisdom of the elders". 
$$
\frac{d^2 X}{dx^2} = k^2 X(x) \implies \boxed{X(x) = Ae^{kx} + Be^{-kx}}

and likewise for $Y$:

$$
\frac{d^2 Y}{dy^2} = k^2 Y(y) \implies Y(y) = \boxed{C\cos(ky) + D\sin(ky)}
$$

Three corollaries:

1. Laplace's eq'n is a linear differential equation. So any linear combination of solutions is a solutions. 
2. $A$, $B$, $C$, $D$, and $k$ are determined/constrained by B.C.'s (see example 3.3)
3. Generally, coef's from linear superposition are determined from integrals (or orthogonality relations, see example 3.3)

## Digression on Completness and Orthogonality
A "complete set" of functions $f_n(y)$ is one that any other function can be written as a linear combination of. \
In math:

$$
f(y) = \sum_{n=0}^\infty
c_n f_n(y)$$

Orthogonal functions obey:

$$
\int_a^y f_n(y)f_{n'}(y) = 0, \quad n \neq n'
$$

e.g., sines:

$$
\int_0^{2\pi} \sin(3x)\sin(4x)dx = 0
$$

The Legendre polynomials are a complete orthogonal set!

$$
f(\theta) = \sum c_\ell P_\ell(\cos \theta) = f(\cos \theta) \quad \text{(completeness)}
$$

$$
\int_{-1}^{1}P_\ell(x)P_{\ell '}(x)dx = \left\{
                \begin{array}{ll}
                  0, \quad \ell \neq \ell'\\
                  \frac{2}{2\ell+1},\quad\ell = 0
                \end{array}
                \right.
$$

Let $x = \cos \theta$:

$$
\frac{dx}{d\theta} = -\sin \theta \implies dx = -\sin \theta d\theta, \, \text{so}
$$

$$
\int_{-1}^{1}P_\ell(x)P_{\ell '}(x)dx \longrightarrow - \int_{\pi}^{0}P_\ell(\cos \theta)P_{\ell '}(\cos \theta)\sin \theta d\theta
$$

$$
=\left\{
                \begin{array}{ll}
                  0&\text{if } \ell \neq \ell'\\
                  \frac{2}{2\ell+1},&\text{if }\ell = \ell'
                \end{array}
                \right.
$$

Or, more conventionally presented:

$$
\boxed{\int_{0}^{\pi}P_\ell(\cos \theta)P_{\ell '}(\cos \theta)\sin \theta d\theta = \left\{
                \begin{array}{ll}
                  0&\text{if } \ell \neq \ell'\\
                  \frac{2}{2\ell+1},&\text{if }\ell = \ell'
                \end{array}
                \right.}
$$
---
Now back to the show:

Completeness of $P_\ell$'s guarantees $f(\cos\theta) = \sum c_\ell P_\ell(\cos\theta)$. Where $c_\ell = A_\ell R_\ell$,we can definitely solve:
$$

Now we must use orthogonality: multiply both sides by $P_\ell' \cos \theta \sin\theta d\theta$:

$$
\begin{align*}
\int_0^\pi V_0(\theta) P_{\ell'}\cos(\theta) \sin (\theta) d \theta &= \int_0^\pi\underbrace{\sum\limits_{\ell = 0}^\infty A_\ell R^\ell P_\ell}_{\text{sum on $A$}}\left(\cos \theta \right) P_{\ell'}\cos(\theta) \sin (\theta) d \theta \\
&= \sum\limits_{\ell = 0}^\infty AR^\ell P_\ell \int_0^\pi P_\ell \left(\cos \theta \right) P_{\ell'}\cos(\theta) \sin (\theta) d \theta \\
&= \sum\limits_{\ell = 0}^\infty A_\ell R^\ell \left(\frac{2}{2 \ell + 1}\right) \quad \text{or } 0 \text{ if } \ell \neq \ell' \\
\implies \int_0^\pi V_0(\theta) P_{\ell'}\cos(\theta) \sin (\theta) d \theta &= A_\ell R^\ell \left(\frac{2}{2 \ell + 1}\right)
\end{align*}
$$

$$
\implies \boxed{A_\ell = \frac{2 \ell + 1}{2R^\ell} \int V_0(\theta) P_\ell (\cos \theta) \sin \theta d \theta}
$$

Cycloid/Cyclotron: Since $\vec E \perp \vec B$, let $\vec E$ be along $\hat z$ and $\vec B$ be along $\hat x$.

Then $$F = q(\underbrace{\vec E}_{\text{along $z$}} + \underbrace{\vec v \times \vec B}_\text{$\perp$ to $\hat x$}) \implies F_x = 0: \quad \text{no component of foce along } x$$

$$
v = \cancel{v_x \hat x}+ v_y \hat y + v_z \hat z
$$

Now, since $v = (0, \dot y, \dot z)$ and $B = (B_x, 0, 0)$, \
$\vec v \times \vec B = B\dot z \hat y = B \dot y \hat z, $ which means

$$
F = q(E \dot z - B \dot z \hat y  - B \dot y \hat z)
$$

thus, 

$$
F_y = m \ddot y = qB \dot z \implies \ddot y = \frac{qB}{m}\dot z = \omega \dot z \quad \text{where} \quad \boxed{\omega = \frac{qB}{m}}
$$

is the **Cyclotron frequency**.

$$

F_z = m &\ddot z = q(E - B \dot y) \\

\end{align*}

Interestingly, there is a force on the wire from the charges moving within it. 

$$
F = q(\vec v \times \vec B) \implies d\vec F = dq(\vec v \times \vec B) \\
\implies \int dF = \int dq(\vec v \times \vec B ) \\
\implies \vec F = \int (\vec v \times \vec B) \lambda d l
$$

or, rather, 

$$
F = \int |(\vec v \times \vec B)| \lambda dl = \int |\lambda \vec v \times \vec B|dl = \int |\vec I \times \vec B| d \vec l

Which is a bit of a swindle, but it's a reasonable one. 

Now to generalize to 2d and 3d:

If $I \sim$ constant along the wire, $\vec F = I \int d \vec l \times \vec B$

* In 1D: $\vec I = \lambda \vec v$ (linear current)
* In 2D: We <u>define</u> $\boxed{\vec K \equiv \frac{d \vec I}{d l_\perp}}$ as the **Surface current density**. \
Note: current/length is curious in terms of units. 

These units suggest that $\boxed{\vec K = \sigma \vec v}$, which is what we <u>use</u>. 

$$
\begin{align*}
|\vec F| = \int d \vec F = \int dq(\nabla \times \vec B) &= \int \sigma da(\vec v \times \vec B) \\
&= \int |\sigma \vec v \times \vec B| d\vec a = \int |\vec K \times \vec B|da \\
\implies \boxed{\vec F = \int \vec K \times \vec B da}
\end{align*}
$$

* In 3d: The analogous quantity is the mobile volume charge density, defined as $\boxed{\vec J \equiv \frac{dI}{da_\perp}}$, THEN $\vec J = \rho \vec v$
and then in 2d: $f = \int dF = \int dq|(\vec v \times \vec B)| = \int \rho |(\vec v \times \vec B)|d\tau$

Now consider $\int \vec J \cdot d \vec a$. By the divergence theorem:

$$
\int\limits_\text{VOL} (\nabla \cdot \vec J) d\tau\quad = \oint\limits_\text{SURFACE} \vec J \cdot d \vec a
$$

Let's explore this conceptually for a second. $[\nabla \cdot \vec J] = \frac{1}{\text{length}} \cdot \frac{\text{current}}{\text{area}} = \frac{\text{charge}}{\text{time} \cdot \text{volume}}$

This all leads to $\boxed{\nabla \cdot \vec J = -\frac{\partial \rho}{\partial t}} \quad \text{The Continuity Equation}$

In words: the time rate of volume charge density is equal to the divergence of the volume.
